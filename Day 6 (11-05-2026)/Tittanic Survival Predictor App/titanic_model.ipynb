```python
import pandas as pd
```

```python
df=pd.read_csv("titanic.csv")
```

```python
df.isnull().sum()   
```

```python
df.drop(columns=["Cabin"], inplace=True)
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.isnull().sum()   
```

```python
df.duplicated().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Encode Embarked (Male=1, Female=0)
le = LabelEncoder()
df["Sex"] = le.fit_transform(df["Sex"])
```

```python
df['Embarked'] = df['Embarked'].map({
    'S': 0,
    'C': 1,
    'Q': 2
})
```

```python
X = df[['Pclass', 'Sex', 'Age', 'Fare', 'SibSp', 'Parch', 'Embarked']]
y = df['Survived']
```

```python
df['Sex']
```

```python
df['Embarked'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=38)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Logistic Regression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Logistic Regression
y_pred_log = log_reg.predict(X_test)
print("\nLogistic Regression Results")
print("Accuracy:", accuracy_score(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log, target_names=["Female","Male"]))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred_log)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Female", "Male"])
disp.plot()
plt.title("Confusion Matrix - Logistic Regression")
plt.show()
```

```python
#'Pclass', 'Sex', 'Age', 'Fare', 'SibSp', 'Parch', 'Embarked'
new_data = [[1, 0, 20, 25.0, 0, 0, 1]]

new_data_df = pd.DataFrame(
    new_data,
    columns=X.columns
)

prediction = log_reg.predict(new_data_df)[0]

if prediction == 1:
    print("Survived")
else:
    print("Not Survived")
```
