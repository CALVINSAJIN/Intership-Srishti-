```python
pp = df.duplicated().value_counts().plot(kind='bar')
plt.title('Number of Duplicated Rows') 
plt.xlabel('Duplicated (False=Unique, True=Duplicate)')
plt.ylabel('Count')
plt.tight_layout()

plt.show()
```

```python
df = df.drop_duplicates()
```

```python
print(df.duplicated().value_counts())
```

```python
print(df.isnull().sum())
```

```python
counts = df.groupby('Harmful to Humans').size()
counts.plot(kind='bar')
plt.title('Harmful Bacteria to Humans')
plt.xlabel('Harmful to Humans')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
counts = df.groupby('Family').size()
plt.scatter(range(len(counts)), counts.values)
plt.title('Bacteria by Family')
plt.xlabel('Family')
plt.ylabel('Count')
plt.xticks(range(len(counts)), counts.index, rotation=45)
plt.tight_layout()
plt.show()
```

```python
counts = df.groupby('Family').size()
plt.pie(counts.values, labels=counts.index, autopct='%1.1f%%')
plt.title('Bacteria by Family')
plt.tight_layout()
plt.show()
```
