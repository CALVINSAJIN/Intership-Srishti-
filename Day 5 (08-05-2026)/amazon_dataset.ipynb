```python
df['rating_count'] = pd.to_numeric(df['rating_count'].str.replace(',', ''), errors='coerce')
df['rating_count'] = df['rating_count'].fillna(df['rating_count'].median()).astype(int)
```

```python
df.isnull().sum()
```

```python
tp=df.groupby('category')['rating_count'].mean().sort_values(ascending=False)
tp.plot(kind='bar', color='orange')
plt.title('Average Rating Count by Category')
plt.ylabel('Average Rating Count')  
plt.xlabel('Category')

plt.tight_layout()
plt.show()
```

```python
# Convert rating column to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Remove invalid values if any
df = df.dropna(subset=['rating'])

labels = ['Rating 0-1', 'Rating 1-2', 'Rating 2-3', 'Rating 3-4', 'Rating 4-5']
rating_bins = [0, 1, 2, 3, 4, 5]

rating_ranges = pd.cut(
    df['rating'],
    bins=rating_bins,
    labels=labels,
    include_lowest=True,
    right=True
)

rating_counts = rating_ranges.value_counts().sort_index()

print('Product counts by rating range:')
display(rating_counts)

plt.figure(figsize=(8, 8))
plt.pie(
    rating_counts,
    labels=rating_counts.index,
    autopct='%1.1f%%',
    colors=sns.color_palette('viridis', len(rating_counts))
)

plt.title('Product Distribution by Rating Range')
plt.tight_layout()
plt.show()
```
